# 1. Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("student-mat.csv", sep=";")
df.head()
```



In this step, the student performance dataset (`student-mat.csv`) is loaded using the Pandas library. The dataset is read using `read_csv()` and semicolon (`;`) is used as the separator.

## Inspect Dataset Information

```python
df.info()
```



The `df.info()` function is used to inspect the dataset structure, including the number of rows, columns, non-null values, and data types of each column.



## Inspect Dataset Shape

```python
print(df.shape)
```



The `df.shape` function is used to check the dimensions of the dataset. It shows the total number of rows and columns present in the dataset.

## Statistical Summary of Dataset

```python
df.describe()
```



The `df.describe()` function is used to generate a statistical summary of the numerical columns in the dataset. It provides information such as count, mean, standard deviation, minimum, maximum, and quartile values.

## Check Missing Values

```python
df.isnull().sum()
```



In this step, missing values are checked using `.isnull().sum()`. This helps identify whether any column contains null or missing data.

## Remove Duplicate Rows

```python
df.drop_duplicates(inplace=True)
print("Duplicate rows removed")
```



The `drop_duplicates()` function is used to remove duplicate rows from the dataset. This helps improve data quality and ensures accurate analysis by avoiding repeated records.

## Average Final Grade

```python
average_grade = df['G3'].mean()
print("Average Final Grade (G3):", average_grade)
```



In this step, the average final grade (`G3`) of all students is calculated using the `mean()` function. This helps understand the overall academic performance of students.

## Students Scoring Above 15

```python
high_scorers = df[df['G3'] > 15].shape[0]
print("Students who scored above 15:", high_scorers)
```



This analysis counts the number of students who scored more than 15 marks in the final grade (`G3`). It helps identify high-performing students in the dataset.

## Study Time and Performance Correlation

```python
correlation = df['studytime'].corr(df['G3'])
print("Correlation between Study Time and Final Grade:", correlation)
```



The correlation between study time and final grades is analyzed to understand whether spending more time studying improves student performance.

## Gender-wise Performance Analysis

```python
gender_performance = df.groupby('sex')['G3'].mean()
print(gender_performance)
```



This step compares the average final grades of male and female students to identify which gender performs better on average.

## Histogram of Grades

```python
plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=10, kde=True)

plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Number of Students')

plt.show()
```



A histogram is used to visualize the distribution of final grades (`G3`) among students. It helps understand how student scores are spread across different grade ranges.

## Scatterplot of Study Time vs Grades

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['studytime'], y=df['G3'])

plt.title('Study Time vs Final Grades')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')

plt.show()
```



A scatterplot is created to analyze the relationship between study time and final grades (`G3`). It helps determine whether increased study time affects academic performance.

## Gender-wise Average Score Comparison

```python
gender_avg = df.groupby('sex')['G3'].mean()

plt.figure(figsize=(6,5))
gender_avg.plot(kind='bar')

plt.title('Average Final Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Grade (G3)')

plt.show()
```



A bar chart is used to compare the average final grades (`G3`) of male and female students. This helps identify differences in academic performance between genders.

## Conclusion

From this analysis, the student performance dataset was loaded, explored, cleaned, analyzed, and visualized. The results show average final grades, high-scoring students, study time relationship with grades, and gender-wise performance comparison.
